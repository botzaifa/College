# Find any missing values 

```python
df.isnull().head()
```

# Filling Null Values

```python
df2 = df.fillna(value = 0)
df2.head()
```

```python
df2.isnull().sum().sum()
```

```python
df3 = df.fillna(value = df['balcony'].mean())
df3.head()
```

```python
df3 = df.fillna(value = df['balcony'].min())
df3.head()
```

```python
df4 = df.fillna(value = df['balcony'].max())
df4.head()
```

# Dropping null values 

```python
df5 = df.dropna()
df5.head()
```

```python
df6 = df.dropna(how = 'any')
df6.head()
```

# Replacing null values 


```python
df7 = df.replace(to_replace=np.nan,value=121)
df7.head()
```
