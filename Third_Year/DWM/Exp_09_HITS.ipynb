```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.DiGraph()
```

```python
g.add_edges_from([ ("A","D"),("B","C"),("B","E"),("C","A"),("D","C"),("E","B"),("E","C"),
        ("E","F"),("E","D"),("F","C"),("F","H"),("G","A"),("G","C"), ("H","C") ])
```

```python
nx.draw_networkx(g, with_labels=True)
```

```python
hub,authority = nx.hits(g, max_iter=50, normalized=True)
```

```python
print(f"Hubs => {hub}")
print(f"Authority Scores => {authority}")
```
