```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Generate a synthetic dataset with 4 clusters
X, _ = make_blobs(n_samples=300, centers=4, random_state=0, cluster_std=1.0)
```

```python
# Create and fit the K-Means model with 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
# Get the cluster centers and labels
cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_
```

```python
# Plot the data points and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1,], s=200, c='red')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering")
plt.show()
```
